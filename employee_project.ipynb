***Label Encoding Technique***

```python
from sklearn.preprocessing import LabelEncoder
encoder_df = LabelEncoder()
df['workclass']= encoder_df.fit_transform(df['workclass'])
df['education']= encoder_df.fit_transform(df['education'])
df['occupation']= encoder_df.fit_transform(df['occupation'])
df['gender']= encoder_df.fit_transform(df['gender'])
df['native-country']= encoder_df.fit_transform(df['native-country'])
```

*Dividing the dataset into input & output variable*

```python
X= df.drop(columns=['income'])            #input variables
Y = df['income']                          #output variables
```

```python
X
```

```python
Y
```

# Feature Scaling 

***Normalisation (Scale the data)***

```python
from sklearn.preprocessing import MinMaxScaler
df_scaler = MinMaxScaler()
X = df_scaler.fit_transform(X)
```

```python
X
```

# Spliting the data

***Train-Test Split***

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=30,stratify=Y)
```

```python
X_train.shape, X_test.shape
```

```python
X_train
```

```python
X_train= pd.DataFrame(X_train)
```

```python
X_train
```

***Applying Machine Learning Algorithm***

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier
df_knn = KNeighborsClassifier()
df_knn.fit(X_train, Y_train)
predict = df_knn.predict(X_test)
predict
```

```python
from sklearn.metrics import accuracy_score          #finding accuracy score
accuracy = accuracy_score(Y_test,predict)
print("Accuracy:", accuracy)
```

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
df_lreg = LogisticRegression()
df_lreg.fit(X_train, Y_train)
predict_2 = df_lreg.predict(X_test)
predict_2
```

```python
from sklearn.metrics import accuracy_score          
accuracy_2 = accuracy_score(Y_test,predict_2)
print("Accuracy:", accuracy_2)
```

```python
# Neural Network Classifier

from sklearn.neural_network import MLPClassifier
df_mlp = MLPClassifier(solver = 'adam', hidden_layer_sizes=(5,2), random_state=2, max_iter=2000)
df_mlp.fit(X_train, Y_train)
predict_3 = df_mlp.predict(X_test)
predict_3
```

```python
from sklearn.metrics import accuracy_score          
accuracy_3 = accuracy_score(Y_test,predict_3)
print("Accuracy:", accuracy_3)
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
df_rfc = RandomForestClassifier()
df_rfc.fit(X_train, Y_train)
predict_4 = df_rfc.predict(X_test)
predict_4
```

```python
from sklearn.metrics import accuracy_score          
accuracy_4 = accuracy_score(Y_test,predict_4)
print("Accuracy:", accuracy_4)
```

```python
# Gradient Boosting Classifier

from sklearn.ensemble import GradientBoostingClassifier
df_gbc = GradientBoostingClassifier()
df_gbc.fit(X_train, Y_train)
predict_5 = df_gbc.predict(X_test)
predict_5
```

```python
from sklearn.metrics import accuracy_score          
accuracy_5 = accuracy_score(Y_test,predict_5)
print("Accuracy:", accuracy_5)
```

```python
# SVM

from sklearn.svm import SVC
Svm_model = SVC(kernel='linear', C=1.0, random_state=42)
Svm_model.fit(X_train, Y_train)
predict_6 = Svm_model.predict(X_test)
predict_6
```

```python
from sklearn.metrics import accuracy_score          
accuracy_6 = accuracy_score(Y_test,predict_6)
print("Accuracy:", accuracy_6)
```

# Result Visualization

```python
#Comparison between models
models = ['KNN', 'Logistic Regression', 'Neural Network', 'Random Forest', 'Gradient Boosting', 'SVM']
accuracies = [accuracy, accuracy_2, accuracy_3, accuracy_4, accuracy_5, accuracy_6]

df_compare = pd.DataFrame({ 'Model': models,'Accuracy': accuracies})
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.barplot(x='Model', y='Accuracy', data=df_compare, palette='viridis')

plt.title('Model Comparison', fontsize=16)
plt.xlabel('Classifier Models')
plt.ylabel('Accuracy Score')
plt.ylim(0, 1) 

plt.tight_layout()
plt.show()

```

***Finding & Storing the best model***

```python
model_dict = {
    'KNN': (df_knn, accuracy),
    'Logistic Regression': (df_lreg, accuracy_2),
    'Neural Network': (df_mlp, accuracy_3),
    'Random Forest': (df_rfc, accuracy_4),
    'Gradient Boosting': (df_gbc, accuracy_5),
    'SVM': (Svm_model, accuracy_6)
}

# Finding the best model based on accuracy
best_model_name, (best_model, best_accuracy) = max(model_dict.items(), key=lambda item: item[1][1])

print(model_dict)
print('\n')
print(f"Best Model: {best_model_name} with Accuracy: {best_accuracy}")
```

```python
# Saving the best model to a file
joblib.dump(best_model, f"{best_model_name}_model.pkl")

print(f"Model saved as {best_model_name}_model.pkl")
```

# Real-Time Salary Prediction Interface

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import joblib

model = joblib.load("Gradient Boosting_model.pkl")

st.markdown("""
    <style>
    .main {background-color: #f0f8ff;}
    h1 {color: #2e8b57;}
    .stButton button {
        background-color: #008cba;
        color: white;
        border-radius: 8px;
        height: 3em;
        width: 100%;
    }
    .stDownloadButton button {
        background-color: #e67300;
        color: white;
        border-radius: 8px;
    }
    .stMarkdown {color: #4b0082;}
    </style>
""", unsafe_allow_html=True)

# App Config
st.set_page_config(page_title="ðŸ’¹ ESC ðŸ’² Salary Classifier", page_icon="ðŸ’¹", layout="centered")

# Title and Tagline
st.title("Employee Salary Classification App")
st.markdown("""
This dazzling app predicts whether an employee earns **>50K or <=50K** annually.
Input data manually or try batch prediction with a CSV â€” now in living color! ðŸŒˆ
""")

# Sidebar Input Form
st.sidebar.header("ðŸŽ¨ Enter Details with Flair")
with st.sidebar.expander("ðŸŒŸ Individual Prediction", expanded=True):
    age = st.slider("ðŸŽ‚ Age", 18, 65, 30)
    education = st.selectbox("ðŸ“š Education", ["Bachelors", "Masters", "Phd", "HS-grad", "Assoc", "some-college"])
    occupation = st.selectbox("ðŸ§‘â€ðŸ”§ Occupation", [
        "Tech-support", "Craft-repair", "Sales", "Exec-managerial", "Prof-speciality",
        "Handlers-cleaners", "Machine-op-inspct", "Adm-clerical", "Farming-fishing",
        "Transport-moving", "Priv-house-serv", "Protective-serv", "Armed-Forces"
    ])
    hours_per_week = st.slider("ðŸ•“ Weekly Hours", 1, 80, 40)
    experience = st.slider("ðŸ“ˆ Experience (Years)", 0, 40, 5)

input_df = pd.DataFrame({
    'age': [age],
    'education': [education],
    'occupation': [occupation],
    'hours-per-week': [hours_per_week],
    'experience': [experience]
})

# Show input data
st.subheader("ðŸ“‹ Your Inputs")
st.dataframe(input_df.style.highlight_max(axis=0, color='lightgreen'))

# Prediction Button
if st.button("ðŸš€ Predict Now", type="primary"):
    prediction = model.predict(input_df)
    st.success(f"ðŸ§  AI thinks: **{prediction[0]}**")

# Batch Prediction Section
st.markdown("---")
st.subheader("ðŸ“‚ CSV Batch Prediction")

uploaded_file = st.file_uploader("ðŸ“ Upload CSV for Bulk Predictions", type=["csv"])
if uploaded_file:
    batch_data = pd.read_csv(uploaded_file)
    st.write("ðŸ•µï¸ Preview Uploaded Data")
    st.dataframe(batch_data.style.highlight_max(axis=0, color='lightblue'))

    batch_preds = model.predict(batch_data)
    batch_data['PredictedClass'] = batch_preds

    st.write("ðŸŽ¯ Batch Predictions")
    st.dataframe(batch_data.style.highlight_max(axis=0, color='lightcoral'))

    csv = batch_data.to_csv(index=False).encode('utf-8')
    st.download_button("ðŸ“¥ Download CSV", csv, file_name="Predicted_Results.csv", mime="text/csv")

st.markdown("ðŸ’¬ Made with ðŸ’– by Tanishq")
```
